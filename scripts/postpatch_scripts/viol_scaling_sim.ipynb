```python
import numpy as np
import random

# see how arbitrage scales with number of questions in tuple
# with multi-paraphrasechecker
# it's linear

N = 500
# random_probs = [random.uniform(0, 1) for _ in range(N)]
random_probs = [1/(1+np.exp(-random.normalvariate(0, 1))) for _ in range(N)]

def viol(*probs):
    probs = np.array(probs)
    n = len(probs)
    return - n * np.log(
        np.prod(probs) ** (1 / n) + np.prod(1 - probs) ** (1 / n)
    )

# plot viol(*random_probs[:x])
import matplotlib.pyplot as plt

x_values = range(1, N + 1)
y_values = [viol(*random_probs[:x]) for x in x_values]

plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('Viol')
plt.title('Viol Scaling Simulation')
plt.show()
```
